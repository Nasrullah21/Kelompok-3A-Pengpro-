## 2.1 DERIVATIVES

#### Turunan suatu fungsi y = f (x) dengan satu variabel x didefinisikan sebagai

$$
\frac{d}{dx} f(x) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x} \tag{2.1.1}
$$

### Definisi Turunan (Persamaan 2.1.1)

Turunan dari fungsi \( f(x) \) terhadap \( x \) didefinisikan sebagai:

$$
\frac{d}{dx} f(x) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x} \tag{2.1.1}
$$

#### Penurunan:

Misalkan kita ingin mengetahui bagaimana nilai \( f(x) \) berubah ketika \( x \) berubah sedikit sebesar \( \Delta x \).

Selisih perubahan fungsi dapat dituliskan sebagai:
$$
\Delta f = f(x + \Delta x) - f(x) \tag{2.1.1.1}
$$

Lalu kita hitung **rata-rata laju perubahan** terhadap \( x \):
$$
\frac{\Delta f}{\Delta x} = \frac{f(x + \Delta x) - f(x)}{\Delta x} \tag{2.1.1.2}
$$

Untuk mendapatkan **laju perubahan tepat di titik \( x \)**, kita ambil limit saat \( \Delta x \to 0 \):

$$
\frac{df}{dx} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x} \tag{2.1.1.3}
$$

Itulah definisi turunan fungsi terhadap \( x \), yang memberi tahu seberapa cepat fungsi berubah pada titik \( x \).


#### Hal ini sering digantikan dengan salah satu perkiraan berikut:

#### üî∏ Keterangan Simbol:

- \( f(x) \) : Fungsi terhadap variabel bebas \( x \)  
- \( $\Delta x$ \) : Perubahan kecil pada variabel \( x \)  
- \( f(x + $\Delta x$) \) : Nilai fungsi saat \( x \) bertambah sebesar \( $\Delta x$ \)  
- \( f(x +$ \Delta x$) - f(x) \) : Perubahan nilai fungsi akibat perubahan \( x \)  

- \( $\displaystyle \frac{df}{dx}$ \) : Turunan fungsi \( f \) terhadap \( x \), menyatakan laju perubahan fungsi secara instan di titik \( x \)


**Perbedaan ke depan:**

$$
\frac{f(x+h) - f(x)}{h} \tag{2.1.2a}
$$

### Bentuk Pendekatan Numerik Turunan: Selisih Maju  
$$
\frac{f(x+h) - f(x)}{h} \tag{2.1.2a}
$$


#### üîπ Penjelasan dan Penurunan:

Bentuk ini merupakan pendekatan numerik untuk turunan fungsi \( f(x) \) di sekitar titik \( x \), menggunakan nilai fungsi di titik \( x \) dan \( x + h \), dengan \( h \) adalah langkah kecil (finite difference).

Jika kita ambil definisi turunan dari:

$$
\frac{df}{dx} = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h} \tag{2.1.2a.1}
$$

Maka bentuk (2.1.2a) adalah bentuk **pendekatannya** untuk \( h \) kecil namun tidak nol:

$$
f'(x) \approx \frac{f(x+h) - f(x)}{h} \tag{2.1.2a.2}
$$

Ini disebut **forward difference approximation** dari turunan pertama fungsi.



#### üî∏ Keterangan Simbol:

- \( f(x) \) : Fungsi terhadap variabel \( x \)  
- \( h \) : Nilai kecil positif (langkah perubahan \( x \))  
- \( f(x+h) \) : Nilai fungsi pada titik yang sedikit lebih besar dari \( x \)  
- \( $ \frac{f(x+h) - f(x)}{h} $\) : Laju perubahan rata-rata fungsi dari \( x \) ke \( x+h \)  
- Digunakan untuk **pendekatan numerik** turunan saat turunan analitik sulit dihitung


#### üîπ Catatan:

- Semakin kecil \( h \), semakin akurat pendekatannya.
- Namun jika \( h \) terlalu kecil, error pembulatan (floating-point error) bisa muncul dalam komputasi numerik.



**Perbedaan mundur:**

$$
\frac{f(x) - f(x - h)}{h} \tag{2.1.2b}
$$

### Pendekatan Numerik Turunan: Selisih Mundur  
$$
\frac{f(x) - f(x - h)}{h} \tag{2.1.2b}
$$

#### üîπ Penjelasan dan Penurunan:

Rumus ini merupakan bentuk pendekatan numerik untuk turunan fungsi \( f(x) \) dengan menggunakan nilai fungsi di titik \( x \) dan titik sebelumnya \( x - h \). Ini disebut **selisih mundur (backward difference)**.

Dari definisi turunan:

$$
\frac{df}{dx} = \lim_{h \to 0} \frac{f(x) - f(x - h)}{h} \tag{2.1.2b.1}
$$

Maka pendekatan numeriknya (jika \( h \) cukup kecil) adalah:

$$
f'(x) \approx \frac{f(x) - f(x - h)}{h} \tag{2.1.2b.2}
$$

#### üî∏ Keterangan Simbol:

- \( f(x) \) : Fungsi terhadap variabel bebas \( x \)  
- \( h \) : Langkah kecil positif pada \( x \)  
- \( f(x - h) \) : Nilai fungsi pada titik sebelum \( x \), yaitu \( x - h \)  
- \( $\dfrac{f(x) - f(x - h)}{h} $\) : Laju perubahan fungsi dihitung mundur dari titik \( x \)  
- Digunakan dalam metode numerik jika data atau informasi tersedia dari arah belakang (ke kiri)



#### üîπ Catatan:

- Digunakan saat data pada \( x + h \) tidak tersedia.
- Akurasinya setara dengan bentuk selisih maju, tapi arah pendekatannya dari kiri.



**Perbedaan sentral:**

$$
\frac{f(x + h) - f(x - h)}{2h} \tag{2.1.2c}
$$

### Pendekatan Numerik Turunan: Selisih Sentral  
$$
\frac{f(x + h) - f(x - h)}{2h} \tag{2.1.2c}
$$


#### üîπ Penjelasan dan Penurunan:

Bentuk ini disebut **selisih sentral (central difference)** karena menggunakan dua titik, yaitu satu di depan dan satu di belakang dari titik \( x \): yaitu \( x + h \) dan \( x - h \).

Turunan didekati dengan:

$$
f'(x) \approx \frac{f(x + h) - f(x - h)}{2h} \tag{2.1.2c.1}
$$

Ini merupakan rata-rata dari pendekatan maju dan mundur:

$$
\text{Central} = \frac{1}{2} \left( \frac{f(x + h) - f(x)}{h} + \frac{f(x) - f(x - h)}{h} \right) \tag{2.1.2c.2}
$$


#### üî∏ Keterangan Simbol:

- \( f(x) \) : Fungsi terhadap variabel bebas \( x \)  
- \( h \) : Langkah kecil positif  
- \( f(x + h) \) : Nilai fungsi sedikit di depan \( x \)  
- \( f(x - h) \) : Nilai fungsi sedikit di belakang \( x \)  
- \( $\dfrac{f(x + h) - f(x - h)}{2h} $\) : Laju perubahan fungsi dihitung berdasarkan selisih dua sisi titik \( x \)  
- Pendekatan ini memiliki **tingkat ketelitian lebih tinggi** dibanding bentuk selisih maju dan mundur (orde error \( O(h^2) \))



#### üîπ Catatan:

- Selisih sentral lebih akurat karena mengurangi error numerik akibat asimetri.
- Umumnya digunakan dalam komputasi numerik seperti metode beda hingga (finite difference method).



#### Turunan parsial dari fungsi multivariabel, misalnya f (x, y) dari dua variabel x dan y, terhadap x dan y, didefinisikan sebagai

$$
\frac{\partial}{\partial x} f(x, y) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x, y) - f(x, y)}{\Delta x} \tag{2.1.3a}
$$

### Definisi Turunan Parsial terhadap \( x \)  
$$
\frac{\partial}{\partial x} f(x, y) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x, y) - f(x, y)}{\Delta x} \tag{2.1.3a}
$$


#### üîπ Penjelasan dan Penurunan:

Turunan parsial digunakan untuk fungsi dua variabel atau lebih.  
Untuk \( f(x, y) \), **turunan parsial terhadap \( x \)** berarti kita menurunkan fungsi terhadap \( x \) dengan **menganggap \( y \) sebagai konstan**.

Dari definisi limit:

$$
\frac{\partial f}{\partial x} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x, y) - f(x, y)}{\Delta x} \tag{2.1.3a.1}
$$

Bentuk ini sangat mirip dengan turunan biasa, hanya saja:

- \( f \) tergantung pada dua variabel, \( x \) dan \( y \),
- Kita hanya mengizinkan \( x \) berubah (sebanyak \($ \Delta x $\)),
- Sedangkan \( y \) tetap konstan selama proses limit.



#### üî∏ Keterangan Simbol:

- \( f(x, y) \) : Fungsi dua variabel, bergantung pada \( x \) dan \( y \)  
- \( $\Delta x$ \) : Perubahan kecil pada variabel \( x \)  
- \( $f(x + \Delta x, y)$ \) : Nilai fungsi saat \( x \) bertambah \( $\Delta x $\), dengan \( y \) tetap  
- \( $\dfrac{f(x + \Delta x, y) - f(x, y)}{\Delta x} $\) : Laju perubahan fungsi terhadap \( x \), saat \( y \) tetap  
- \($ \displaystyle \frac{\partial f}{\partial x} $\) : Turunan parsial fungsi terhadap \( x \)  
- \( $\displaystyle \lim_{\Delta x \to 0} $\) : Limit saat perubahan \( x \) mendekati nol



#### üîπ Catatan:

- Proses ini analog dengan turunan biasa, tapi berlaku untuk fungsi multivariabel.
- Jika kita ingin menurunkan terhadap \( y \), maka \( x \) dianggap konstan (lihat persamaan 2.1.3b).



$$
\frac{\partial}{\partial y} f(x, y) = \lim_{\Delta y \to 0} \frac{f(x, y + \Delta y) - f(x, y)}{\Delta y} \tag{2.1.3b}
$$

### Definisi Turunan Parsial terhadap \( y \)  
$$
\frac{\partial}{\partial y} f(x, y) = \lim_{\Delta y \to 0} \frac{f(x, y + \Delta y) - f(x, y)}{\Delta y} \tag{2.1.3b}
$$

#### üîπ Penjelasan dan Penurunan:

Turunan parsial terhadap \( y \) berarti kita melihat bagaimana fungsi \( f(x, y) \) berubah jika hanya variabel \( y \) yang berubah, sedangkan variabel \( x \) **dianggap konstan**.

Rumus ini adalah bentuk definisi limit untuk turunan parsial:

$$
\frac{\partial f}{\partial y} = \lim_{\Delta y \to 0} \frac{f(x, y + \Delta y) - f(x, y)}{\Delta y} \tag{2.1.3b.1}
$$

Proses ini sama seperti turunan biasa, hanya saja:
- Fungsi bergantung pada dua variabel: \( x \) dan \( y \),
- Kita hanya membiarkan \( y \) yang berubah sedikit,
- Sedangkan \( x \) tetap nilainya.

---

#### üî∏ Keterangan Simbol:

- \( $f(x, y)$ \) : Fungsi dua variabel, bergantung pada \( x \) dan \( y \)  
- \( $\Delta y $\) : Perubahan kecil pada variabel \( y \)  
- \( $f(x, y + \Delta y) $\) : Nilai fungsi saat \( y \) bertambah sebesar \( \Delta y \), dengan \( x \) tetap  
- \( $\dfrac{f(x, y + \Delta y) - f(x, y)}{\Delta y}$ \) : Laju perubahan fungsi terhadap \( y \) saat \( x \) tetap  
- \( $\displaystyle \frac{\partial f}{\partial y} $ \) : Turunan parsial fungsi terhadap \( y \)  
- \( $\displaystyle \lim_{\Delta y \to 0} $\) : Limit ketika perubahan \( y \) mendekati nol



#### üîπ Catatan:

- Turunan parsial sangat berguna dalam kalkulus multivariat dan fisika, seperti untuk gradien, divergensi, dan hukum konservasi.
- Jika kita menurunkan terhadap \( x \), maka gunakan definisi serupa dengan \( \Delta x \), seperti pada (2.1.3a).



### \<Aturan Rantai\>

Misalkan \( y = f(u) \) merupakan fungsi diferensial dari variabel \( u \) di mana \( u = g(x) \) merupakan fungsi dari variabel lain \( x \). Maka *aturan rantai* menyatakan bahwa turunan dari \( y \) terhadap \( x \) adalah

$$
\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx} \tag{2.1.4}
$$

Jika $ z = f(u, v) $ merupakan fungsi diferensiabel dari dua variabel \( u \) dan \( v \) di mana \( u = g(x, y) \) dan \( v = h(x, y) \) merupakan fungsi dari dua variabel lain \( x \) dan \( y \), dan keduanya mempunyai turunan parsial kontinu, maka *aturan rantai* menyatakan bahwa turunan parsial dari \( z \) terhadap \( x \) dan \( y \) adalah

$$
\frac{\partial z}{\partial x} = \frac{\partial z}{\partial u} \cdot \frac{\partial u}{\partial x} + \frac{\partial z}{\partial v} \cdot \frac{\partial v}{\partial x}, dan
\frac{\partial z}{\partial y} = \frac{\partial z}{\partial u} \cdot \frac{\partial u}{\partial y} + \frac{\partial z}{\partial v} \cdot \frac{\partial v}{\partial y} \tag{2.1.5}
$$

### \<Gradien\>

*Gradien* (1D) dari fungsi satu variabel $ f(x) $ dapat didefinisikan sebagai

$$
\vec{\nabla} f(x) = \vec{u}_x \frac{df}{dx}
$$

*Gradien* (2D) dari fungsi dua variabel $ f(x, y) $ didefinisikan sebagai

$$
\vec{\nabla} f(x, y) = \vec{u}_x \frac{\partial f}{\partial x} + \vec{u}_y \frac{\partial f}{\partial y} \tag{2.1.7}
$$

*Gradien* (3D) dari fungsi tiga variabel $ f(x, y, z) $ didefinisikan sebagai

$$
\vec{\nabla} f(x, y, z) = \vec{u}_x \frac{\partial f}{\partial x} + \vec{u}_y \frac{\partial f}{\partial y} + \vec{u}_z \frac{\partial f}{\partial z} \tag{2.1.8}
$$

#### di mana $ \vec{\nabla} $ (disebut ‚Äúdel‚Äù atau ‚Äúnabla‚Äù) menunjukkan *operator diferensial vektor* (didefinisikan dalam sistem koordinat Cartesian 3D) sebagai

$$
\vec{\nabla} = \vec{u}_x \frac{\partial}{\partial x} + \vec{u}_y \frac{\partial}{\partial y} + \vec{u}_z \frac{\partial}{\partial z} \tag{2.1.9}
$$



dengan $ \vec{u}_x, \vec{u}_y, \vec{u}_z $ menjadi vektor satuan sepanjang sumbu x, y, dan z, berturut-turut.

### \<Directional Derivative\>

#### *Turunan arah* dari \( f(x, y, z) \) dalam arah vektor satuan  $ \vec{u} = \alpha_x \vec{u}_x + \alpha_y \vec{u}_y + \alpha_z \vec{u}_z $, di mana $ \|\vec{u}\| = 1 $, didefinisikan sebagai:

$$
\vec{\nabla}_u f(x, y, z) = \lim_{t \to 0} \frac{f(x + t\alpha_x, y + t\alpha_y, z + t\alpha_z) - f(x, y, z)}{t}
$$

$$
\frac{\partial f}{\partial x}(x + t\alpha_x) \cdot \frac{dt}{dt} +
\frac{\partial f}{\partial y}(y + t\alpha_y) \cdot \frac{dt}{dt} +
\frac{\partial f}{\partial z}(z + t\alpha_z) \cdot \frac{dt}{dt} =
\vec{\nabla} f(x, y, z) \cdot \vec{u} \tag{2.1.10}
$$

#### Garis/Sumbu Singgung (Tangent Line/Plane)
Persamaan *garis singgung* dari sebuah kurva yang dideskripsikan oleh \( y = f(x) \) pada titik \( (x_1, y_1) \) dapat dituliskan sebagai:



$$
y = y_1 + \left. \frac{\partial f}{\partial x} \right|_{x = x_1} (x - x_1) \tag{2.1.11}
$$

Persamaan *garis singgung* dari kurva level yang dijelaskan oleh $f(x, y) = c$ di titik $(x_1, y_1)$ adalah:


$$
\left. \frac{\partial f}{\partial x} \right|_{(x_1, y_1)} (x - x_1) +
\left. \frac{\partial f}{\partial y} \right|_{(x_1, y_1)} (y - y_1) = 0 \tag{2.1.12}
$$

Persamaan *bidang singgung* dari permukaan level yang dijelaskan oleh $f(x, y, z) = c$ di titik $(x_1, y_1, z_1)$ adalah:


$$
\left. \frac{\partial f}{\partial x} \right|_{(x_1, y_1, z_1)} (x - x_1) +
\left. \frac{\partial f}{\partial y} \right|_{(x_1, y_1, z_1)} (y - y_1) +
\left. \frac{\partial f}{\partial z} \right|_{(x_1, y_1, z_1)} (z - z_1) = 0 \tag{2.1.13}
$$

Persamaan (2.1.12) dan (2.1.13) dapat diturunkan dari Persamaan (1.1.39) dengan gradien sebagai vektor normal terhadap bidang singgung. Persamaan (2.1.11) identik dengan Persamaan (2.1.12) dengan f(x)-y=0 menggantikan f (x, y )=c.

Contoh 2.2 Gradien Negatif dari Potensial Listrik untuk Mendapatkan Medan Listrik

Potensial listrik \( V \) yang dibentuk oleh dua muatan titik positif/negatif \( +Q / -Q \) (dengan besar yang sama dan polaritas berlawanan), masing-masing terletak di \( (0, +d/2, 0) \) dan \( (0, -d/2, 0) \), diberikan oleh:

$$
V = \frac{Q}{4\pi \varepsilon_0} \left( \frac{1}{R_1} - \frac{1}{R_2} \right)
= \frac{Q}{4\pi \varepsilon_0}
\left( \frac{1}{\sqrt{x^2 + (y - d/2)^2}} -
\frac{1}{\sqrt{x^2 + (y + d/2)^2}} \right)
\tag{E2.2.1}
$$

Medan listrik E dapat diperoleh sebagai gradien negatif dari V :

$$
\mathbf{E} = -\nabla V(x, y)
= -\mathbf{u}_x \frac{\partial V}{\partial x}
- \mathbf{u}_y \frac{\partial V}{\partial y}
\tag{2.1.7}
$$
Sehingga,

$$
\mathbf{E}(x, y) = \frac{Q}{4\pi \varepsilon_0}
    \left\{
    \frac{x}{\left(x^2 + (y - d/2)^2 \right)^{3/2}} +
    \frac{x}{\left(x^2 + (y + d/2)^2 \right)^{3/2}}
    \right\} \mathbf{u}_x \\
    +\left\{
    \frac{y - d/2}{\left(x^2 + (y - d/2)^2 \right)^{3/2}} -
    \frac{y + d/2}{\left(x^2 + (y + d/2)^2 \right)^{3/2}}
    \right\} \mathbf{u}_y 
    \tag{E2.2.2}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Fungsi potensial listrik (Electric Potential)
def Vxy(x, y, Q, d):
    return Q / np.sqrt(x**2 + (y - d/2)**2 + 1e-8) - Q / np.sqrt(x**2 + (y + d/2)**2 + 1e-8)

# Grid setup
dx = 0.1
dy = 0.1
x = np.arange(-2 + dx/2, 2 + dx/2, dx)
y = np.arange(-2 + dy/2, 2 + dy/2, dy)
X, Y = np.meshgrid(x, y)

# Parameter dan potensial
Q = 1
d = 2
V = Vxy(X, Y, Q, d)

# Komponen medan listrik (dari turunan potensial)
R1 = np.sqrt(X**2 + (Y - d/2)**2) + 1e-8
R2 = np.sqrt(X**2 + (Y + d/2)**2) + 1e-8
Ex = X / R1**3 - X / R2**3
Ey = (Y - d/2) / R1**3 - (Y + d/2) / R2**3

# Normalisasi untuk mengatur panjang panah
a = 1
Ex = np.sign(Ex) * (1 - np.exp(-np.abs(Ex) / a))
Ey = np.sign(Ey) * (1 - np.exp(-np.abs(Ey) / a))

# Plotting
fig = plt.figure(figsize=(12, 5))

# Subplot 1: Surface plot of potential V(x, y)
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(X, Y, V, cmap='viridis', edgecolor='none')
ax1.set_title('Electric Potential $V(x,y)$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('V')
ax1.view_init(elev=15, azim=10)

# Subplot 2: Quiver plot of electric field
ax2 = fig.add_subplot(1, 2, 2)
ax2.quiver(X, Y, Ex, Ey, color='g', scale=20)
ax2.set_title('Electric Field $\mathbf{E} = -\\nabla V$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.axis('equal')

plt.tight_layout()
plt.show()
```

## 2.2  VECTOR FUNCTIONS

Fungsi **vektor (bernilai vektor)** 2D dan 3D dituliskan sebagai:

$$
\mathbf{p}(t) = x(t)\mathbf{u}_x + y(t)\mathbf{u}_y \tag{2.2.1a}
$$

dan

$$
\mathbf{p}(t) = x(t)\mathbf{u}_x + y(t)\mathbf{u}_y + z(t)\mathbf{u}_z \tag{2.2.1b}
$$

masing-masing. Kurva tersebut dapat digambar sebagai kurva 2D dan 3D, mungkin dengan kepala panah yang menunjukkan arah pergerakan titik **p**(t) saat parameter t meningkat. 
Garis singgung pada kurva (yang menggambarkan **p**(t)) pada titik **p**(t1) adalah garis melalui titik **p**(t1) yang sejajar dengan turunan, yang disebut vektor singgung,

$$
\mathbf{p}'(t) = \frac{d\mathbf{p}(t)}{dt} = \mathbf{p}(t) = x(t)\mathbf{u}_x + y(t)\mathbf{u}_y = \frac{dx(t)}{dt}\mathbf{u}_x + \frac{dy(t)}{dt}\mathbf{u}_y = x'(t)\mathbf{u}_x + y'(t)\mathbf{u}_y \tag{2.2.2a}
$$

$$
\mathbf{p}'(t) = \frac{d\mathbf{p}(t)}{dt} = \mathbf{p}(t) = x(t)\mathbf{u}_x + y(t)\mathbf{u}_y + z(t)\mathbf{u}_z =  \frac{dx(t)}{dt}\mathbf{u}_x + \frac{dy(t)}{dt}\mathbf{u}_y + \frac{dz(t)}{dt}\mathbf{u}_z = x'(t)\mathbf{u}_x + y'(t)\mathbf{u}_y + z'(t)\mathbf{u}_z \tag{2.2.2b}
$$

Panjang busur kurva (yang mewakili **p**(t)) dari **p**(t0) dan **p**(t1) dapat diperoleh dengan mengintegralkan besarnya turunan (2.2.2) atas [t0, t1] sebagai

$$
l = \int_{t_0}^{t_1} \|\mathbf{p}'(t)\| dt = \int_{t_0}^{t_1} \sqrt{(x'(t))^2 + (y'(t))^2 + (z'(t))^2} dt \tag{2.2.3}
$$

hanya jika **p**(t) dan kurva yang sesuai bersifat halus dalam artian bahwa fungsi komponen x(t), y(t), dan z(t) memiliki turunan pertama kontinu yang tidak semuanya nol untuk interval terbuka (t0, t1). 
Perhatikan bahwa fungsi vektor **p**(t) dapat diparameterisasi sebagai **p**(l) dalam hal panjang busur

$$
l = \int_{0}^{l} \|\mathbf{p}'(\tau)\| d\tau = \int_{0}^{l} \sqrt{(x'(\tau))^2 + (y'(\tau))^2 + (z'(\tau))^2} d\tau = h(t) \tag{2.2.4}
$$

dengan mensubstitusikan fungsi invers $ t=h^1(l) $ dari $ l=h(t) $ untuk t ke dalam Persamaan (2.2.1), seperti yang diilustrasikan dalam Contoh 
2.3(c). 

Sebelum mengakhiri bagian ini, mari kita perhatikan aturan berikut tentang diferensiasi fungsi vektor yang dapat didiferensiasikan **p**(t), **p**1(t), dan **p**2(t) di mana q(t) adalah fungsi skalar yang dapat didiferensiasikan:

$$
\frac{d}{dt} (q(t)\mathbf{p}(t)) = \left(\frac{d}{dt} q(t)\right)\mathbf{p}(t) + q(t) \left(\frac{d}{dt} \mathbf{p}(t)\right) = q'(t)\mathbf{p}(t) + q(t)\mathbf{p}'(t) \tag{2.2.5a}
$$

$$
\frac{d}{dt} (\mathbf{p}_1(t) + \mathbf{p}_2(t)) = \frac{d}{dt} \mathbf{p}_1(t) + \frac{d}{dt} \mathbf{p}_2(t) = \mathbf{p}_1'(t) + \mathbf{p}_2'(t) \tag{2.2.5b}
$$

$$
\frac{d}{dt} (\mathbf{p}_1(t) \cdot \mathbf{p}_2(t)) = \frac{d}{dt} \mathbf{p}_1(t) \cdot \mathbf{p}_2(t) + \mathbf{p}_1(t) \cdot \frac{d}{dt} \mathbf{p}_2(t) = \mathbf{p}_1'(t) \cdot \mathbf{p}_2(t) + \mathbf{p}_1(t) \cdot \mathbf{p}_2'(t) \tag{2.2.5c}
$$

$$
\frac{d}{dt} (\mathbf{p}_1(t) \times \mathbf{p}_2(t)) = \frac{d}{dt} \mathbf{p}_1(t) \times \mathbf{p}_2(t) + \mathbf{p}_1(t) \times \frac{d}{dt} \mathbf{p}_2(t) = \mathbf{p}_1'(t) \times \mathbf{p}_2(t) + \mathbf{p}_1(t) \times \mathbf{p}_2'(t) \tag{2.2.5d}
$$

```python
import numpy as np
import sympy as sp
from scipy.integrate import quad

# Inisialisasi simbol
t, s, tau = sp.symbols('t s tau', real=True)

# 1. Definisi fungsi vektor heliks p(t)
p_t = sp.Matrix([
    t,
    sp.cos(2 * sp.pi * t),
    sp.sin(2 * sp.pi * t)
])

# 2. Turunan p(t) terhadap t
dp_dt = p_t.diff(t)

# 3. Evaluasi p(t) dan dp/dt di t = 1
p_1 = p_t.subs(t, 1)
dp_dt_1 = dp_dt.subs(t, 1)

# 4. Garis singgung di t = 1
tangent_line = p_1 + (t - 1) * dp_dt_1

# 5. Hitung panjang busur dari t = 0 ke t = 1
# Norm dari dp/dt
dp_dt_norm = sp.sqrt(dp_dt.dot(dp_dt))
dp_dt_norm_func = sp.lambdify(t, dp_dt_norm, 'numpy')
arc_length, _ = quad(dp_dt_norm_func, 0, 1)

# 6. Parameterisasi ulang: t sebagai fungsi s
sqrt_term = sp.sqrt(1 + 4 * sp.pi**2)
t_s = s / sqrt_term

# 7. Bentuk p(s)
p_s = p_t.subs(t, t_s)

# 8. Turunan p(s) terhadap s
dp_ds = p_s.diff(s)

# 9. Tampilkan hasil
print("=== Contoh 2.3 ===")
print("\np(t) =")
sp.pprint(p_t)

print("\nTurunan dp/dt =")
sp.pprint(dp_dt)

print("\np(1) =")
sp.pprint(p_1)

print("\ndp/dt @ t = 1 =")
sp.pprint(dp_dt_1)

print("\nGaris singgung di t = 1:")
sp.pprint(tangent_line)

print(f"\nPanjang busur s dari t = 0 ke t = 1 ‚âà {arc_length:.4f}")

print("\nt(s) =")
sp.pprint(t_s)

print("\np(s) =")
sp.pprint(p_s)

print("\ndp/ds =")
sp.pprint(dp_ds)
```

## 2.3 Velocity and Acceleration

#### Jika posisi partikel atau benda pada waktu t dijelaskan oleh:

##### its first and second derivatives:

$$
\vec{v}(t) = \frac{d}{dt} \vec{p}(t) = \frac{dx(t)}{dt}\vec{u}_x + \frac{dy(t)}{dt}\vec{u}_y + \frac{dz(t)}{dt}\vec{u}_z = x'(t)\vec{u}_x + y'(t)\vec{u}_y + z'(t)\vec{u}_z \tag{2.3.2}
$$

$$
\vec{a}(t) = \frac{d^2}{dt^2} \vec{p}(t) = \frac{d^2x(t)}{dt^2}\vec{u}_x + \frac{d^2y(t)}{dt^2}\vec{u}_y + \frac{d^2z(t)}{dt^2}\vec{u}_z = x''(t)\vec{u}_x + y''(t)\vec{u}_y + z''(t)\vec{u}_z \tag{2.3.3}
$$

#### disebut kecepatan dan percepatan partikel, masing-masing. Besarnya kecepatan

$$
\|\vec{v}(t)\| = \||\vec  p '(t) || = \sqrt{\left(\frac{dx(t)}{dt}\right)^2 + \left(\frac{dy(t)}{dt}\right)^2 + \left(\frac{dz(t)}{dt}\right)^2} = \sqrt{(x'(t))^2 + (y'(t))^2 + (z'(t))^2} \tag{2.3.4}
$$

#### Perhatikan bahwa jika kecepatannya konstan maka

$$
\|\vec{v}\|^2 = \vec{v} \cdot \vec{v} = c^2 \tag{2.3.6}
$$

#### Ini menyiratkan bahwa vektor kecepatan dan vektor percepatan partikel yang bergerak dengan kecepatan konstan saling tegak lurus. Untuk partikel yang bergerak dengan kecepatan konstan sepanjang lingkaran (pada bidang xy) dengan jari-jari R, vektor posisinya dijelaskan oleh

$$
\vec{p}(t) = R\cos(\omega t)\vec{u}_x + R\sin(\omega t)\vec{u}_y \tag{2.3.8}
$$

Dengan demikian kita dapat menggunakan Persamaan (2.3.2) dan (2.3.3) untuk mendapatkan vektor kecepatan dan percepatan (sentripetal) sebagai berikut:

$$
\mathbf{v}(t){=}
\frac{d}{dt} \mathbf{p}(t) =
-R\omega \sin(\omega t) \, \mathbf{u}_x
+ R\omega \cos(\omega t) \, \mathbf{u}_y
\qquad \tag {2.3.9}
$$

$$
\mathbf{a}(t) \overset{\text{(2.3.3)}}{=}
\frac{d\mathbf{v}(t)}{dt}^{\text{(2.3.9)}}
= -R\omega^2 \cos(\omega t)\, \mathbf{u}_x
- R\omega^2 \sin(\omega t)\, \mathbf{u}_y
\overset{\text{(2.3.8)}}{=}
-\omega^2 \mathbf{p}(t)
\qquad \tag {2.3.10}
$$


##### Di sisi lain, kita dapat mengintegrasikan percepatan a(t) secara berurutan untuk mendapatkan kecepatan v(t) dan vektor posisi p(t):

$$
\vec{v}(t) = \int_{-\infty}^t \vec{a}(t) dt = \int_{0}^t \vec{a}(t) dt + \int_{-\infty}^{0} \vec{a}(t) dt = \vec{v}(t_0) + \int_{t_0}^{t} \vec{a}(t) dt \tag{2.3.12}
$$

$$
\vec{p}(t) = \int_{-\infty}^{t} \vec{v}(t) dt = \int_{0}^{t} \vec{v}(t) dt + \int_{-\infty}^{0} \vec{v}(t) dt = \vec{p}(t_0) + \int_{t_0}^{t} \vec{v}(t) dt \tag{2.3.13}
$$

Sekarang, sebagai jumlah penyimpangan kurva halus yang mewakili **p**(t) dari datar atau lurus, kita mendefinisikan kelengkungan kurva pada suatu titik sebagai turunan vektor tangen satuan **u**T  w.r.t. terhadap parameter panjang busur *l*:

$$ 
\kappa = \left|\frac{d\mathbf u_T}{dl}\right| = 
\left|\frac{d \mathbf u_T/dt}{dl/dt}\right| = 
\frac{| \mathbf u_T'(t)|}{|| l'(t)|| } =
\frac{|| \mathbf u_T'(t)||}{||\mathbf p'(t)||=v(t)} 
\tag{2.3.14}
$$

di mana vektor tangen satuan **u**T didefinisikan sebagai

$$ 
\mathbf u_T = 
\frac{\mathbf v(t)}{||\mathbf v(t)||=v(t)} = 
\frac{\mathbf p'(t)}{||\mathbf p'(t)||}=
\frac{d\mathbf p(t) /dt} {dl(t) /dt} =
\frac{d\mathbf p(l)} {dl}
\tag {2.3.15} $$

Kelengkungan yang didefinisikan oleh Persamaan (2.3.14) menggambarkan seberapa tiba-tiba/halusnya suatu kurva mengubah arahnya sehubungan dengan panjang busur pada suatu titik **p**(t) pada kurva. Selain itu, jari-jari kelengkungan didefinisikan sebagai kebalikan dari kelengkungan:

$$ 
r_k = \frac{1}{\kappa} \tag {2.3.16}
$$

Lingkaran yang melalui suatu titik p(t) dengan jari-jari $ r_k $ yang paling sesuai dengan kurva disebut lingkaran kelengkungan atau lingkaran oskulasi dan pusatnya adalah pusat kelengkungan.

Contoh 2.5 Kelengkungan Elips

Temukan nilai kelengkungan dan lingkaran kelengkungan elips yang direpresentasikan oleh

$$
\mathbf{p}(t) = 2\cos(\pi t)\mathbf{u}_x + \sin(\pi t)\mathbf{u}_y \tag{E2.5.1}
$$

Kami menerapkan Persamaan (2.3.15) untuk Persamaan (E2.5.1) guna menemukan vektor tangen satuan

$$
\mathbf{u}_T = \frac{\mathbf{p}'(t)}{\|\mathbf{p}'(t)\|}
= \frac{-2\pi \sin(\pi t)\mathbf{u}_x + \pi \cos(\pi t)\mathbf{u}_y}{\sqrt{4\pi^2 \sin^2(\pi t) + \pi^2 \cos^2(\pi t)}}

$$
\mathbf{u}_T =
\begin{cases}
\mathbf{u}_y, & \text{pada } t = 0 \\

Kemudian kita menggunakan Persamaan (2.3.14) untuk menemukan kelengkungan

$$
\kappa(t) = \frac{\|\mathbf{u}_T'(t)\|}{\|\mathbf{p}'(t)\|}
= \frac  {2\pi^2 \cos(\pi t)\mathbf{u}_x - \pi^2 \sin(\pi t)\mathbf{u}_y}{\sqrt{4\pi^2 \sin^2(\pi t) + \pi^2 \cos^2(\pi t)}}
\kappa(t) =
\begin{cases}
2, & \text{pada } t = 0 \\
1/2, & \text{pada } t = 1/2
\end{cases}
\tag{E2.5.3}
$$

Kita dapat menjalankan skrip MATLAB di atas em02e05.m untuk memplot dua lingkaran kelengkungan bersama-sama dengan elips seperti yang digambarkan pada Gambar 2.5, di mana kedua lingkaran tersebut direpresentasikan oleh

$$
x^2 + (y + 1)^2 = 1/(1/2)^2 \quad \text{with radius } r_k1 = 1/\kappa(0) = 1/2 \text{ passing through } (2, 0)
\tag{E2.5.4a}
$$

$$
(x - 1)^2 + y^2 = 1/(2)^2 \quad \text{with radius } r_k1 = 1/\kappa(1/2) = 2 \text{ passing through } (0, 1)
\tag{E2.5.4b}
$$

Di sini, sama seperti Persamaan (2.3.7), dengan memperhatikan bahwa besarnya vektor tangen satuan **u**T(t) adalah konstan sehingga

$$
\|\mathbf{u}_T(t)\|^2 = \mathbf{u}_T \cdot \mathbf{u}_T = 1
$$

Diketahui:
$$
\|\mathbf{u}_T(t)\|^2 = \mathbf{u}_T(t) \cdot \mathbf{u}_T(t) = 1
$$

Turunkan kedua ruas terhadap \( t \):
$$
\frac{d}{dt}(\mathbf{u}_T \cdot \mathbf{u}_T) = \frac{d}{dt}(1)
$$

Gunakan aturan turunan dari dot product:
$$
\frac{d}{dt}(\mathbf{u}_T \cdot \mathbf{u}_T) = \frac{d\mathbf{u}_T}{dt} \cdot \mathbf{u}_T + \mathbf{u}_T \cdot \frac{d\mathbf{u}_T}{dt}
= 2 \frac{d\mathbf{u}_T}{dt} \cdot \mathbf{u}_T
$$

Karena turunan dari konstanta adalah nol:
$$
2 \frac{d\mathbf{u}_T}{dt} \cdot \mathbf{u}_T = 0
\quad \Rightarrow \quad
\frac{d\mathbf{u}_T}{dt} \cdot \mathbf{u}_T = 0
$$

Artinya:
> Turunan dari vektor tangen \( \mathbf{u}_T \) ortogonal terhadap dirinya sendiri.


kita dapat membedakan ini w.r.t. *t* untuk mendapatkan

$$
\mathbf{u}_T(t) \cdot \mathbf{u}_T'(t) = 0
\tag{2.3.17}
$$

Hal ini menyiratkan bahwa **u** T(t) dan **u'** T(t) saling tegak lurus. Dengan demikian, kita dapat mendefinisikan vektor normal satuan terhadap kurva, yang tegak lurus terhadap vektor tangen **u**T(t), sebagai

$$
\mathbf{u}_N(t) = \frac{\mathbf{u}_T'(t)}{\|\mathbf{u}_T'(t)\|}
\tag{2.3.18}
$$

Dari Persamaan (2.3.15) juga dapat dituliskan kecepatannya sebagai

$$
\mathbf{v}(t) = v(t) \mathbf{u}_T(t)
\tag{2.3.19}
$$

Membedakan hal ini sehubungan dengan t melalui Persamaan (2.2.5a) menghasilkan ekspresi percepatan sebagai

$$
\mathbf{a}(t) = \frac{d}{dt} \mathbf{v}(t)
= \frac{d}{dt} \left( v(t)\mathbf{u}_\mathrm{T}(t) \right) =
q'(t)\mathbf{p}(t) + q(t)\mathbf{p}'(t)
= v'(t)\mathbf{u}_\mathrm{T}(t) + v(t)\mathbf{u}_\mathrm{T}'(t)
= v'(t)\mathbf{u}_\mathrm{T}(t) + v(t)\left\| \mathbf{u}_\mathrm{T}'(t) \right\|\mathbf{u}_\mathrm{N}(t)
$$

$$

= v'(t)\mathbf{u}_\mathrm{T}(t) + \kappa(t)v^2(t)\mathbf{u}_\mathrm{N}(t)
= a_\mathrm{T} \mathbf{u}_\mathrm{T}(t) + a_\mathrm{N} \mathbf{u}_\mathrm{N}(t)
\quad \text{with } a_\mathrm{T} = v'(t) \text{ and } a_\mathrm{N} = \kappa(t)v^2(t)
\qquad \tag {2.3.20}
$$

Di sini, *aT* dan *aN*, yang disebut komponen tangensial dan normal (atau sentripetal) dari percepatan, masing-masing mencerminkan perubahan vektor kecepatan **v**(t) dalam besaran dan arah. 
Sehubungan dengan vektor tangen satuan **u** T(t) dan vektor normal satuan **u**N(t), kita mendefinisikan vektor satuan ketiga, yang disebut vektor binormal, sebagai

$$
\mathbf{u}_B = \mathbf{u}_T \times \mathbf{u}_N
\tag{2.3.21}
$$

Ketiga vektor **u**T, **u**N, dan **u**B membentuk sistem koordinat tangan kanan dari vektor-vektor yang saling ortogonal, yang disebut kerangka Frenet-Serret atau segitiga bergerak. Bidang yang direntangkan oleh **u**T dan **u**N, bidang yang direntangkan oleh **u**N dan **u**B, dan bidang yang direntangkan oleh **u**B dan **u**T masing-masing disebut bidang oskulasi, bidang normal, dan bidang penyearah. Bidang-bidang tersebut diilustrasikan dalam Gambar 2.6.

```python
import numpy as np
import matplotlib.pyplot as plt

# Domain waktu
t = np.linspace(0, 1, 1000)

# Vektor posisi p(t)
x = 2 * np.cos(np.pi * t)
y = np.sin(np.pi * t)

# Turunan pertama p'(t)
dx_dt = -2 * np.pi * np.sin(np.pi * t)
dy_dt = np.pi * np.cos(np.pi * t)

# Kecepatan v(t)
v = np.sqrt(dx_dt**2 + dy_dt**2)

# Vektor tangen satuan u_T(t)
u_T_x = dx_dt / v
u_T_y = dy_dt / v

# Turunan dari u_T(t) terhadap t
du_T_x_dt = np.gradient(u_T_x, t)
du_T_y_dt = np.gradient(u_T_y, t)

# Besar dari du_T/dt
norm_du_T_dt = np.sqrt(du_T_x_dt**2 + du_T_y_dt**2)

# Kelengkungan: kappa(t) = ||du_T/dt|| / v
kappa = norm_du_T_dt / v

# Titik-titik khusus
t0 = 0
t1 = 0.5

# Evaluasi pada titik t = 0 dan t = 0.5
i0 = np.argmin(np.abs(t - t0))
i1 = np.argmin(np.abs(t - t1))

# Output hasil numerik
print("Hasil evaluasi:")
print(f"u_T(0) = ({u_T_x[i0]:.2f}, {u_T_y[i0]:.2f})")
print(f"u_T(0.5) = ({u_T_x[i1]:.2f}, {u_T_y[i1]:.2f})")
print(f"kappa(0) = {kappa[i0]:.2f}")
print(f"kappa(0.5) = {kappa[i1]:.2f}")
print(f"1/kappa(0) = {1/kappa[i0]:.2f} (radius of curvature)")
print(f"1/kappa(0.5) = {1/kappa[i1]:.2f} (radius of curvature)")

# Plot kurva dan vektor tangen
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='p(t)')
plt.quiver(x[i0], y[i0], u_T_x[i0], u_T_y[i0], scale=5, color='r', label='u_T(0)')
plt.quiver(x[i1], y[i1], u_T_x[i1], u_T_y[i1], scale=5, color='g', label='u_T(0.5)')
plt.scatter([x[i0], x[i1]], [y[i0], y[i1]], color='k')
plt.text(x[i0], y[i0], 't=0', fontsize=10, verticalalignment='bottom')
plt.text(x[i1], y[i1], 't=0.5', fontsize=10, verticalalignment='top')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contoh 2.5: Vektor Tangen Satuan dan Kelengkungan')
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```

**Frenet-Serret formulas**

Sebelum mengakhiri bagian ini, ada baiknya untuk melihat rumus Frenet‚ÄìSerret, yang dapat ditulis dalam bentuk matriks sebagai

$$
\begin{bmatrix}
\mathbf d {u}_T /dl \\
\mathbf d {u}_N /dl \\
\mathbf d {u}_B /dl
\end{bmatrix}
=
\begin{bmatrix}
0 & \kappa & 0 \\
-\kappa & 0 & \tau \\
0 & -\tau & 0
\end{bmatrix}
\begin{bmatrix}
\mathbf{u}_T \\
\mathbf{u}_N \\
\mathbf{u}_B
\end{bmatrix}
\tag{2.3.22}
$$

di mana  $\tau$ = $ \mathbf -(d {u}_B/dl) $ adalah konstanta proporsionalitas negatif antara $ \mathbf -(d {u}_B/dl) $ dan **u**N, disebut torsi kurva, yang mengukur seberapa banyak kurva dengan lilitan trihedral yang bergerak (dari planar) saat parameter t meningkat. 

(Bukti) 

Pertama, kita kalikan Persamaan (2.3.14) dan (2.3.18) untuk mendapatkan Persamaan (2.3.22a):

$$
\kappa \mathbf{u}_N = \frac{\|\mathbf{u}_T'(t)\|}{\|\mathbf{p}'(t)\|} \frac{\mathbf{u}_T'}{\|\mathbf{u}_T'\|}
= \frac{d\mathbf{u}_T/dt}{\|\mathbf{p}'(t)\|} = \frac{d\mathbf{u}_T}{dl}
\tag{2.3.23}
$$

Kemudian, seperti pada Persamaan (2.3.17), perhatikan bahwa

$$
\mathbf{u}_B \cdot \frac{d\mathbf{u}_B}{dl} = 0
\tag{2.3.24}
$$

hanya karena **u**B juga merupakan vektor satuan dengan besaran konstan. Sekarang, diferensiasikan **u**B=**u**T x **u**N(Persamaan (2.3.21)) sehubungan dengan parameter panjang busur l dan menggunakan Persamaan (2.2.5d) menghasilkan

$$
\frac{d\mathbf{u}_B}{dl} = \frac{d}{dl} (\mathbf{u}_T \times \mathbf{u}_N)
= \frac{d\mathbf{u}_T}{dl} \times \mathbf{u}_N + \mathbf{u}_T \times \frac{d\mathbf{u}_N}{dl}
\tag{2.3.25}
$$

## salah cari kode benar

Persamaan (2.3.24) dan (2.3.25) menyiratkan bahwa $ \mathbf -(d {u}_B/dl) $ tegak lurus terhadap **u**B dan **u**T sehingga harus berada dalam arah **u**N. Itulah sebabnya kita dapat menulis Persamaan (2.3.22c): 

$$
\frac{d\mathbf{u}_B}{dl} = -\tau \mathbf{u}_N 
\quad \text{with torsion } \tau = -\frac{d\mathbf{u}_B}{dl} \cdot \mathbf{u}_N
\tag{2.3.26}
$$

Sekarang, diferensiasikan N B T u u u w.r.t. l dan menggunakan persamaan (2.2.5d) menghasilkan 

$$
\frac{d\mathbf{u}_\mathrm{N}}{dl}=
\frac{d\mathbf{u}_\mathrm{B}}{dl} \times \mathbf{u}_\mathrm{T}
+ \mathbf{u}_\mathrm{B} \times
\frac{d\mathbf{u}_\mathrm{T}}{dl} =
-\tau \mathbf{u}_\mathrm{N} \times \mathbf{u}_\mathrm{T}
+ \mathbf{u}_\mathrm{B} \times \kappa \mathbf{u}_\mathrm{N}
= \tau \mathbf{u}_\mathrm{B} - \kappa \mathbf{u}_\mathrm{T}
\qquad \tag {2.3.27}
$$

## 2.4 DIVERGENCE AND CURL

#### Divergensi medan vektor F (dijelaskan oleh fungsi vektor dua atau tiga variabel) pada titik pengamatan didefinisikan sebagai

$$
\text{div} \, \mathbf{F} = \lim_{\Delta V \to 0} \frac{1}{\Delta V} \iint_{\Delta S} \mathbf{F} \cdot d\mathbf{s}
\quad 
\tag{2.4.1}
$$

Keterangan:

* dS = u‚Çô √ó dS, yaitu vektor permukaan kecil dengan arah keluar
* ŒîV = volume kecil yang mengelilingi titik pengamatan
* F ¬∑ dS = hasil dot product antara medan vektor dan elemen permukaan

**Interpretasi:** Divergence mengukur seberapa besar aliran keluar dari suatu titik per satuan volume saat volume menyusut ke nol.

Curl

Curl dari medan vektor **F** pada suatu titik didefinisikan sebagai:

$$
\text{curl} \, \mathbf{F} = \lim_{\Delta S \to 0} \frac{1}{\Delta S} \oint_{C} \mathbf{F} \cdot d\mathbf{l} \times \mathbf{u}_n \tag{2.4.2}
$$

Keterangan:

* ‚àÆ F ¬∑ dl = integral garis tertutup
* ŒîS = luas kecil dari permukaan datar yang dibatasi oleh lintasan tertutup C
* u‚Çô = vektor satuan normal dari permukaan, sesuai aturan tangan kanan

**Interpretasi:** Curl mengukur seberapa besar rotasi (putaran) medan vektor di sekitar titik tersebut.


$$
\mathbf{F}(x, y, z) = F_x(x, y, z)\,\mathbf{i} + F_y(x, y, z)\,\mathbf{j} + F_z(x, y, z)\,\mathbf{k}
$$

**Gradien dari fungsi skalar** $f(x, y, z)$:

$$
\nabla f = \frac{\partial f}{\partial x}\,\mathbf{i} + \frac{\partial f}{\partial y}\,\mathbf{j} + \frac{\partial f}{\partial z}\,\mathbf{k} \tag{2.4.3a}
$$

**Divergence dari vektor** $\mathbf{F}$:

$$
\nabla \cdot \mathbf{F} = \frac{\partial F_x}{\partial x} + \frac{\partial F_y}{\partial y} + \frac{\partial F_z}{\partial z} \tag{2.4.4a}
$$

**Penjabaran:**

* Ambil turunan parsial dari komponen Fx terhadap x
* Ambil turunan parsial Fy terhadap y
* Ambil turunan parsial Fz terhadap z
* Jumlahkan ketiganya


* Divergence mengukur apakah suatu titik bersifat sebagai sumber atau penyerap aliran.
* Curl mengukur tingkat perputaran lokal dari medan vektor.
* Konsep ini sangat penting dalam fisika, seperti dalam medan elektromagnetik dan fluida dinamis.

#### (1) Sistem koordinat kartesius dengan f (x,y,z) dan F = Fxux + Fyuy + Fzuz

- **Gradient**:
  $$
  \nabla f = \frac{\partial f}{\partial x} \mathbf{u}_x + \frac{\partial f}{\partial y} \mathbf{u}_y + \frac{\partial f}{\partial z} \mathbf{u}_z \quad \tag{2.4.3a}
  $$

- **Divergence**:
  $$
  \nabla \cdot \mathbf{F} = \frac{\partial F_x}{\partial x} + \frac{\partial F_y}{\partial y} + \frac{\partial F_z}{\partial z} \quad \tag{2.4.4a}
  $$

- **Curl**:
  $$
  \nabla \times \mathbf{F} =
  \left( \frac{\partial F_z}{\partial y} - \frac{\partial F_y}{\partial z} \right) \mathbf{u}_x +
  \left( \frac{\partial F_x}{\partial z} - \frac{\partial F_z}{\partial x} \right) \mathbf{u}_y +
  \left( \frac{\partial F_y}{\partial x} - \frac{\partial F_x}{\partial y} \right) \mathbf{u}_z
  \quad \tag{2.4.5a}
  $$

- **Laplacian**:
  $$
  \nabla \cdot \nabla f = \nabla^2 f =
  \frac{\partial^2 f}{\partial x^2} + \frac{\partial^2 f}{\partial y^2} + \frac{\partial^2 f}{\partial z^2}
  \quad \tag{2.4.6a}
  $$

### (2) Sistem koordinat silinder dengan f (U ,I ,z) dan F = FU ux + FI uI + Fzuz

- **Gradient**:
  $$
  \nabla f = \frac{\partial f}{\partial \rho} \mathbf{u}_\rho + \frac{1}{\rho} \frac{\partial f}{\partial \phi} \mathbf{u}_\phi + \frac{\partial f}{\partial z} \mathbf{u}_z
  \quad \tag{2.4.3b}
  $$

- **Divergence**:
  $$
  \nabla \cdot \mathbf{F} = \frac{1}{\rho} \frac{\partial (\rho F_\rho)}{\partial \rho} + \frac{1}{\rho} \frac{\partial F_\phi}{\partial \phi} + \frac{\partial F_z}{\partial z}
  \quad \tag{2.4.4b}
  $$

- **Curl**:
  $$
  \nabla \times \mathbf{F} = 
  \left( \frac{1}{\rho} \frac{\partial F_z}{\partial \phi} - \frac{\partial F_\phi}{\partial z} \right) \mathbf{u}_\rho +
  \left( \frac{\partial F_\rho}{\partial z} - \frac{\partial F_z}{\partial \rho} \right) \mathbf{u}_\phi +
  \frac{1}{\rho} \left( \frac{\partial (\rho F_\phi)}{\partial \rho} - \frac{\partial F_\rho}{\partial \phi} \right) \mathbf{u}_z
  \quad \tag{2.4.5b}
  $$

- **Laplacian**:
  $$
  \nabla^2 f = \frac{1}{\rho} \frac{\partial}{\partial \rho} \left( \rho \frac{\partial f}{\partial \rho} \right) + \frac{1}{\rho^2} \frac{\partial^2 f}{\partial \phi^2} + \frac{\partial^2 f}{\partial z^2}
  \quad \tag{2.4.6b}
  $$

#### (3) Sistem koordinat bola dengan f (r,T , I ) dan F = Frur + FT uT + FI uI

- **Gradient**:
  $$
  \nabla f = \frac{\partial f}{\partial r} \mathbf{u}_r + \frac{1}{r} \frac{\partial f}{\partial \theta} \mathbf{u}_\theta + \frac{1}{r \sin \theta} \frac{\partial f}{\partial \phi} \mathbf{u}_\phi
  \quad \tag{2.4.3c}
  $$

- **Divergence**:
  $$
  \nabla \cdot \mathbf{F} =
  \frac{1}{r^2} \frac{\partial (r^2 F_r)}{\partial r} +
  \frac{1}{r \sin \theta} \frac{\partial (F_\theta \sin \theta)}{\partial \theta} +
  \frac{1}{r \sin \theta} \frac{\partial F_\phi}{\partial \phi}
  \quad \tag{2.4.4c}
  $$

- **Curl**:
  $$
  \nabla \times \mathbf{F} =
  \frac{1}{r \sin \theta} \left( \frac{\partial (F_\phi \sin \theta)}{\partial \theta} - \frac{\partial F_\theta}{\partial \phi} \right) \mathbf{u}_r +
  \frac{1}{r} \left( \frac{1}{\sin \theta} \frac{\partial F_r}{\partial \phi} - \frac{\partial (r F_\phi)}{\partial r} \right) \mathbf{u}_\theta +
  \frac{1}{r} \left( \frac{\partial (r F_\theta)}{\partial r} - \frac{\partial F_r}{\partial \theta} \right) \mathbf{u}_\phi
  \quad \tag{2.4.5c}
  $$

- **Laplacian**:
  $$
  \nabla \cdot \nabla f = \nabla^2 f =
  \frac{1}{r^2} \frac{\partial}{\partial r} \left( r^2 \frac{\partial f}{\partial r} \right)
  + \frac{1}{r^2 \sin \theta} \frac{\partial}{\partial \theta} \left( \sin \theta \frac{\partial f}{\partial \theta} \right)
  + \frac{1}{r^2 \sin^2 \theta} \frac{\partial^2 f}{\partial \phi^2}
  \tag{2.4.6c}
  $$


Rumus-rumus ini telah dimasukkan ke dalam fungsi-fungsi MATLAB berikut:

```python
from sympy import symbols, diff, sin, simplify, pprint

def diva(F, coords='car'):
    """
    Menghitung divergensi analitik dari fungsi vektor 3D simbolik F.
    
    Parameters:
        F : tuple/list dengan 3 elemen (fungsi simbolik)
        coords : 'car' (default), 'cyl', atau 'sph' (untuk koordinat Kartesian, Silinder, atau Bola)
    
    Returns:
        divF : hasil divergensi simbolik
    """
    coords = coords.lower()[:2]  # ambil dua huruf pertama
    if coords == 'cy':  # Koordinat silinder
        r, ph, z = symbols('r ph z')
        divF = diff(r * F[0], r) / r + diff(F[1], ph) / r + diff(F[2], z)  # Eq (2.4.4b)
        coord_name = 'cylindrical'
    elif coords == 'sp':  # Koordinat bola
        r, th, ph = symbols('r th ph')
        divF = (
            diff(r**2 * F[0], r) / r**2 +
            diff(F[1] * sin(th), th) / (r * sin(th)) +
            diff(F[2], ph) / (r * sin(th))
        )  # Eq (2.4.4c)
        coord_name = 'spherical'
    else:  # Default: koordinat Kartesian
        x, y, z = symbols('x y z')
        divF = diff(F[0], x) + diff(F[1], y) + diff(F[2], z)  # Eq (2.4.4a)
        coord_name = 'Cartesian'

    # Tampilkan jika tidak dipakai sebagai fungsi return
    print(f"Div(F) in {coord_name} coordinates is:")
    pprint(simplify(divF))

    return simplify(divF)
```

```python
from sympy import symbols, diff, sin, simplify, pprint

def curla(F, coords='car'):
    """
    Menghitung curl analitik dari fungsi vektor 3D simbolik F.
    
    Parameters:
        F : list atau tuple dari 3 komponen simbolik
            Untuk:
            - 'car' -> [Fx, Fy, Fz]
            - 'cyl' -> [Frho, Fphi, Fz]
            - 'sph' -> [Fr, Ftheta, Fphi]
        coords : 'car', 'cyl', atau 'sph' (default = 'car')
    
    Returns:
        curlF : vektor hasil curl (dalam bentuk simbolik)
    """
    coords = coords.lower()[:2]
    if coords == 'cy':  # Silinder (Eq. 2.4.5b)
        r, ph, z = symbols('r ph z')
        coord_name = 'cylindrical'
        curlF = [
            diff(F[2], ph)/r - diff(F[1], z),
            diff(F[0], z) - diff(F[2], r),
            (diff(r * F[1], r) - diff(F[0], ph)) / r
        ]
    elif coords == 'sp':  # Bola (Eq. 2.4.5c)
        r, th, ph = symbols('r th ph')
        coord_name = 'spherical'
        curlF = [
            (diff(F[2] * sin(th), th) - diff(F[1], ph)) / (r * sin(th)),
            (diff(F[0], th) / sin(ph) - diff(r * F[2], r)) / r,
            (diff(r * F[1], r) - diff(F[0], th)) / r
        ]
    else:  # Kartesian (Eq. 2.4.5a)
        x, y, z = symbols('x y z')
        coord_name = 'Cartesian'
        curlF = [
            diff(F[2], y) - diff(F[1], z),
            diff(F[0], z) - diff(F[2], x),
            diff(F[1], x) - diff(F[0], y)
        ]
    
    # Tampilkan hasil
    print(f"Curl(F) in {coord_name} coordinates is:")
    for comp in curlF:
        pprint(simplify(comp))
    
    return [simplify(c) for c in curlF]
```

```python
from sympy import symbols, diff, sin, simplify, pprint

def gradienta(f, coords='car'):
    """
    Menghitung gradien dari fungsi skalar simbolik f
    dalam sistem koordinat: 'car', 'cyl', atau 'sph'.

    Parameters:
        f : fungsi simbolik skalar
        coords : string (default 'car') ‚Üí 'car', 'cyl', 'sph'

    Returns:
        Df : vektor gradien simbolik
    """
    coords = coords.lower()[:2]
    
    if coords == 'cy':  # Silinder (Eq. 2.4.3b)
        r, ph, z = symbols('r ph z')
        coord_name = 'cylindrical'
        Df = [
            diff(f, r),
            diff(f, ph) / r,
            diff(f, z)
        ]
    elif coords == 'sp':  # Bola (Eq. 2.4.3c)
        r, th, ph = symbols('r th ph')
        coord_name = 'spherical'
        Df = [
            diff(f, r),
            diff(f, th) / r,
            diff(f, ph) / (r * sin(th))
        ]
    else:  # Kartesian (Eq. 2.4.3a)
        x, y, z = symbols('x y z')
        coord_name = 'Cartesian'
        Df = [
            diff(f, x),
            diff(f, y),
            diff(f, z)
        ]
    
    # Tampilkan jika tidak dipakai sebagai nilai kembalian
    print(f"Gradient of f in {coord_name} coordinates is:")
    pprint([simplify(df) for df in Df])
    
    return [simplify(df) for df in Df]
```

```python
from sympy import symbols, diff, sin, simplify, pprint

def laplaciana(f, coords='car'):
    """
    Menghitung Laplacian analitik dari fungsi skalar simbolik f.
    
    Parameters:
        f : fungsi simbolik skalar
        coords : 'car', 'cyl', atau 'sph' (default = 'car')
    
    Returns:
        lapf : hasil Laplacian (simbolik)
    """
    coords = coords.lower()[:2]
    
    if coords == 'cy':  # Koordinat silinder (Eq. 2.4.6b)
        r, ph, z = symbols('r ph z')
        coord_name = 'cylindrical'
        lapf = (
            diff(r * diff(f, r), r) / r +
            diff(diff(f, ph), ph) / r**2 +
            diff(diff(f, z), z)
        )
        
    elif coords == 'sp':  # Koordinat bola (Eq. 2.4.6c)
        r, th, ph = symbols('r th ph')
        coord_name = 'spherical'
        lapf = (
            diff(r**2 * diff(f, r), r) / r**2 +
            diff(sin(th) * diff(f, th), th) / (r**2 * sin(th)) +
            diff(diff(f, ph), ph) / (r**2 * sin(th)**2)
        )
        
    else:  # Default: Kartesian (Eq. 2.4.6a)
        x, y, z = symbols('x y z')
        coord_name = 'Cartesian'
        lapf = diff(f, x, x) + diff(f, y, y) + diff(f, z, z)

    # Tampilkan hasil jika tidak dipakai sebagai nilai kembalian
    print(f"Laplacian(f) in {coord_name} coordinate is:")
    pprint(simplify(lapf))
    
    return simplify(lapf)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, simplify, pretty, init_printing
from sympy.vector import CoordSys3D
init_printing()

# --- Fungsi bantu analitik ---
def diva(F):
    x, y, z = symbols('x y z')
    return simplify(diff(F[0], x) + diff(F[1], y) + diff(F[2], z))

def curla(F):
    x, y, z = symbols('x y z')
    curl = [
        diff(F[2], y) - diff(F[1], z),
        diff(F[0], z) - diff(F[2], x),
        diff(F[1], x) - diff(F[0], y)
    ]
    return [simplify(c) for c in curl]

# --- Grid ---
dx, dy = 0.5, 0.5
xx = np.arange(-2, 2.5, dx)
yy = np.arange(-2, 2.5, dy)
X, Y = np.meshgrid(xx, yy)

# --- Loop untuk dua kasus vektor ---
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
fig.suptitle("Vector Fields and Their Divergence/Curl", fontsize=14)

x, y, z = symbols('x y z')

for iter in range(2):
    if iter == 0:
        FX = X**3
        FY = Y**3
        Fx, Fy, Fz = x**3, y**3, 0
        div_analytical = 3 * (X**2 + Y**2)
        curl_analytical = np.zeros_like(X)
    else:
        FX = -Y**3
        FY = X**3
        Fx, Fy, Fz = -y**3, x**3, 0
        div_analytical = np.zeros_like(X)
        curl_analytical = 3 * (X**2 + Y**2)

    # Numerik Divergence & Curl (2D)
    div_numeric = np.gradient(FX, dx, axis=1) + np.gradient(FY, dy, axis=0)
    curl_numeric = np.gradient(FY, dx, axis=1) - np.gradient(FX, dy, axis=0)

    # --- Plot Vektor ---
    ax1 = axes[iter, 0]
    ax1.quiver(X, Y, FX, FY)
    ax1.set_title(f'F{iter+1}(x,y)')
    ax1.axis('equal')

    # --- Plot Divergence ---
    ax2 = axes[iter, 1]
    im2 = ax2.imshow(div_numeric, extent=[-2, 2, -2, 2], origin='lower')
    ax2.set_title('div F (numerik)')
    fig.colorbar(im2, ax=ax2)

    # --- Plot Curl ---
    ax3 = axes[iter, 2]
    im3 = ax3.imshow(curl_numeric, extent=[-2, 2, -2, 2], origin='lower')
    ax3.set_title('curl F (numerik)')
    fig.colorbar(im3, ax=ax3)

    # --- Analitik (Symbolic) ---
    print(f"\n===== Vektor F{iter+1}(x,y) =====")
    print("Analitik Divergence:")
    pretty(diva([Fx, Fy, Fz]), use_unicode=True)
    print("Analitik Curl:")
    curlF = curla([Fx, Fy, Fz])
    for i, comp in enumerate(['i', 'j', 'k']):
        print(f"{comp}-komponen: {pretty(curlF[i], use_unicode=True)}")

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

Perhitungan ini dapat dilakukan dengan menggunakan 'diva()' dan 'curla()' sebagai berikut:

```python
from sympy import symbols, simplify, pprint
from sympy import diff

# --- Symbolic variables ---
r, J0, a, ph, z = symbols('r J0 a ph z')

# --- Vector fields in cylindrical coordinates ---
Ha = [0, J0 * r**2 / (2 * a), 0]   # Eq.(2.10.1a)
Hb = [0, J0 * a**2 / r,       0]   # Eq.(2.10.1b)

# --- Divergence function in cylindrical coords (Eq. 2.4.4b) ---
def diva_cyl(F):
    Fr, Fph, Fz = F
    divF = diff(r * Fr, r) / r + diff(Fph, ph) / r + diff(Fz, z)
    return simplify(divF)

# --- Curl function in cylindrical coords (Eq. 2.4.5b) ---
def curla_cyl(F):
    Fr, Fph, Fz = F
    curl_r = diff(Fz, ph)/r - diff(Fph, z)
    curl_ph = diff(Fr, z) - diff(Fz, r)
    curl_z = (diff(r * Fph, r) - diff(Fr, ph)) / r
    return [simplify(curl_r), simplify(curl_ph), simplify(curl_z)]

# --- HITUNG DIVERGENSI ---
print("=== Divergence ===")
print("div(Ha):")
pprint(diva_cyl(Ha))  # Should be J0/a
print("div(Hb):")
pprint(diva_cyl(Hb))  # Should be 0

# --- HITUNG CURL ---
print("\n=== Curl ===")
print("curl(Ha):")
pprint(curla_cyl(Ha))  # Should be [0, 0, J0/a * r]
print("curl(Hb):")
pprint(curla_cyl(Hb))  # Should be [0, 0, 0]
```

\int_C \frac{2(x\mathbf{u}_x + y\mathbf{u}_y + z\mathbf{u}_z)}{(x^2+y^2+z^2)^{3/2}} \,\cdot dl \tag {E2.16.1}
$$

sepanjang tiga kurva $C_1, C_2,$ dan $C_3$ yang digambarkan pada Gambar 2.15 dan diparameterisasi oleh

$$ C_1: (1-t)\mathbf{u}_x + (1-t)\mathbf{u}_y \quad \text{for } 0 \le t \le \frac{1}{2} $$
$$ C_2: t\mathbf{u}_x + (1-t)\mathbf{u}_y \quad \text{for } 0 \le t \le \frac{1}{2} $$
$$ C_3: (1-t)\mathbf{u}_x + t\mathbf{u}_z \quad \text{for } 0 \le t \le \frac{1}{2} \tag {E2.16.2} $$

masing-masing.

Integralnya dihitung sebagai berikut:

$$
\int_C \frac{2(xu_x+yu_y+zu_z)}{(x^2+y^2+z^2)^{3/2}} \cdot dl = \int_0^{1/2} \frac{2((1-t)u_x+0u_y+(1-t)u_z)}{((1-t)^2+0^2+(1-t)^2)^{3/2}} \cdot ((1-t)'u_x+(1-t)'u_z) dt
$$

$$
= \int_0^1 \frac{2((1-t)u_x+(1-t)u_z)}{((1-t)^2+(1-t)^2)^{3/2}} \cdot (-u_x-u_z) dt = \int_0^1 \frac{2 \cdot 2(-u_x-u_z)}{((1-t)^2 \cdot 2)^{3/2}} dt = \int_0^1 \frac{4(-1)}{(1-t)^3 \sqrt{2}} dt = \sqrt{2} - \sqrt{2} \quad \tag {E2.16.3}
$$


$$
\int_{C_2} \frac{2(x \mathbf{u}x + y \mathbf{u}_y + z \mathbf{u}_z)}{(x^2+y^2+z^2)^{3/2}} \cdot d \mathbf{l} = \int_0^{1/2}  \frac{2((1-t) \mathbf{u}x + 0 \mathbf{u}_y + (1-t) \mathbf{u}_z)}{(1^2+0^2+(1-t)^2)^{3/2}} \cdot ((1-t)' \mathbf{u}_z) dt = \int{0}^{1/2} \frac{2(t-1)}{(1^2+(1-t)^2)^{3/2}} dt
$$


$$
=\int_{\pi/4}^{\tan^{-1} 0.5} \frac{-2 \tan \theta}{(1+\tan^2 \theta)^{3/2}} (-\sec^2 \theta) d \theta \stackrel{(E2.14.4)}{=} -2\left(\frac{1}{\sqrt{5}} - \frac{1}{\sqrt{2}}\right) \quad \tag {E2.16.4}
$$

$$
\int_{C_3} \frac{2(x \mathbf{u}x + y \mathbf{u}_y + z \mathbf{u}_z)}{(x^2+y^2+z^2)^{3/2}} \cdot d \mathbf{l} = \int_0^{ \, 1/2} \frac{2((1-t) \mathbf{u}x + 0 \mathbf{u}_y + 0.5 \mathbf{u}_z)}{(1-t)^2+0^2+0.5^2)^{3/2}} \cdot ((1-t)' \mathbf{u}_x) dt = \int{0}^{1/2} \frac{2(t-1)}{((1-t)^2+0.5^2)^{3/2}} dt
$$

$$
=\int_{\tan^{-1} 2}^{\pi/4} \frac{-2 \times 0.5 \tan \theta}{0.5^3(1+\tan^2 \theta)^{3/2}} (-0.5 \sec^2 \theta) d \theta {=} -4\left(\frac{1}{\sqrt{5}} - \frac{1}{\sqrt{2}}\right) \quad \tag {E2.16.5}
$$

## 2.5.2 Kemandirian Jalur

Fungsi vektor **G** dikatakan konservatif jika dapat dituliskan sebagai gradien dari fungsi skalar, misalnya f (x, y, z), yang disebut fungsi potensial untuk **G**:

$$
G(x, y, z) = G_x(x,y,z)u_x + G_y(x,y,z)u_y + G_z(x,y,z)u_z = \frac{\partial f}{\partial x}u_x + \frac{\partial f}{\partial y}u_y + \frac{\partial f}{\partial z}u_z = \nabla f(x,y,z)
\tag {2.5.6}
$$ 

Medan vektor konservatif disebut medan vektor gradien. Bagaimana kita dapat mengetahui apakah medan vektor konservatif atau tidak? Kriteria untuk medan vektor konservatif adalah bahwa $G_x, G_y, G_z,$ dan turunan parsial pertamanya bersifat kontinu, dan selain itu, ikal fungsi vektor **G** adalah nol di daerah terhubung sederhana terbuka R (yang terhubung dan tidak mengandung lubang di mana **G** bersifat singular):

$$
\nabla \times G = \left(\frac{\partial G_z}{\partial y} - \frac{\partial G_y}{\partial z}\right)u_x + \left(\frac{\partial G_x}{\partial z} - \frac{\partial G_z}{\partial x}\right)u_y + \left(\frac{\partial G_y}{\partial x} - \frac{\partial G_x}{\partial y}\right)u_z = 0  :irrotational \tag {2.5.7}
$$

Perhatikan bahwa $\nabla \times \nabla f = 0$ dan dalam kasus fungsi vektor 2D, kriteria (2.5.7) menjadi

$$
\frac{\partial G_y}{\partial x} = \frac{\partial G_x}{\partial y} \tag {2.5.8}
$$

Di sisi lain, integral garis dari satu titik ke titik lain dikatakan bebas lintasan atau 
bebas lintasan jika nilainya hanya bergantung pada dua titik ujung, tetapi tidak pada lintasan 
integrasi. 

Menurut teorema dasar untuk integral garis, jika dan hanya jika **G** adalah medan vektor konservatif **G**=‚àáf dengan fungsi potensial skalar *f*, maka integral garisnya bebas lintasan, i.e

$\int_C G(x,y,z) \cdot dl = \int_C \nabla f(x,y,z) \cdot dl = f(B) - f(A)$ with $A = l(a)$ and $B = l(b)$ 
$$ \tag {2.5.9} $$

di mana $C$ adalah lintasan integrasi yang diparameterisasi oleh $l(t) = x(t)u_x + y(t)u_y + z(t)u_z$ untuk $a \le t \le b$ di daerah terhubung sederhana terbuka $R$. Dengan kata lain, untuk medan vektor konservatif $G$ di atas daerah terhubung sederhana terbuka $R$, integral garisnya di sekitar kurva tertutup sederhana $C$ di $R$, yang disebut *sirkulasi* $G$ di sekitar $C$, adalah nol:

$\oint_C G(x,y,z) \cdot dl = 0$ for any simple closed curve $C$ 
$$ \tag {2.5.10} $$

Perhatikan bahwa suatu daerah $R$ dikatakan *terbuka* jika tidak memiliki batas, dan dikatakan *tertutup* jika daerah tersebut mencakup semua batasnya, jika ada. Suatu daerah $R$ dikatakan *terhubung* jika dua titik di $R$ dapat dihubungkan dengan lintasan yang terletak sepenuhnya di $R$. Suatu kurva $C$ dikatakan *sederhana* jika tidak memotong dirinya sendiri.

Contoh 2.18 Kemandirian Lintasan Integral Garis

Tentukan kemandirian lintasan dari integral garis berikut dalam kaitannya dengan apakah fungsi vektor integran bersifat konservatif atau tidak:

(a) 
$$
\int_C G(x,y) \cdot dl = \int_C \{G_x(x,y)u_x + G_y(x,y)u_y\} \cdot dl = \int_C \frac{xu_x + yu_y}{x^2+y^2} \cdot dl
\tag {E2.18a.1}
$$ 

Fungsi vektor $G(x, y)$ memenuhi kriteria (2.5.8) untuk medan vektor konservatif 2D:

$\frac{\partial G_y}{\partial x} = \frac{\partial}{\partial x}\left(\frac{y}{x^2+y^2}\right) = \frac{-2xy}{(x^2+y^2)^2}$ dan $\frac{\partial G_x}{\partial y} = \frac{\partial}{\partial y}\left(\frac{x}{x^2+y^2}\right) = \frac{-2xy}{(x^2+y^2)^2}$ $$\tag {E2.18a.2}$$

Akan tetapi, faktanya adalah integral garis dari **G**(x, y) tidak bebas lintasan, karena integral garis sepanjang lingkaran satuan (yang merupakan kurva tertutup) tidak sama dengan nol, seperti yang ditemukan dalam Contoh 2.17. Bagaimana kontradiksi yang tampak ini dapat dijelaskan? Masalahnya adalah lubang singularitas **G** pada (0, 0), yang membuat domain terbuka D (yang berisi lintasan integrasi) tidak dapat dihubungkan secara sederhana (lihat Gambar 2.16(a)).

(b) $$\int G(x,y) \cdot dl = \int \{G_x(x,y)u_x + G_y(x,y)u_y\} \cdot dl = \int \frac{-yu_x + xu_y}{x^2 + y^2} \cdot dl \tag {2.18b.1} $$

Fungsi vektor G(x, y) memenuhi kriteria (2.5.8) untuk medan vektor konservatif 2D:

$$\frac{\partial G_y}{\partial x} = \frac{-1}{x^2} = \frac{\partial G_x}{\partial y} \tag {E2.18b.2}
$$

atas dua daerah terhubung sederhana $\{x>0\}$ dan $\{x<0\}$ yang dipisahkan oleh $x=0$ (lihat Gambar 2.16(b)). Fungsi potensial skalar $f$ sehingga $\nabla f = G$ dapat ditemukan sebagai

$f(x,y) = \frac{1}{2} \ln(x^2 + y^2) + C$ with $\frac{\partial f}{\partial x} = \frac{x}{x^2 + y^2}$ and $\frac{\partial f}{\partial y} = \frac{y}{x^2 + y^2}$ $$ \tag {E2.18b.3}$$

Mari kita mengevaluasi integral pada dua jalur:

$$ C_1: \{tu_x + (1-t)u_y \text{ for } 0 \le t \le 2\} and 
\begin{bmatrix}
C_2: \{(1+t)u_x + u_y \text{ for } 0 \le t \le 1\} \\
C_3: \{2u_x + (1-t)u_y \text{ for } 0 \le t \le 1\} \\
C_4: \{(2-t)u_x - u_y \text{ for } 0 \le t \le 1\}
\end{bmatrix}
$$

$$\int_{C_1} \frac{-yu_x + xu_y}{x^2} \cdot d\mathbf{l} = \int_0^2 \frac{-(1-t)u_x + u_y}{1^2} \cdot \{(1-t)'u_y\}dt = \int_0^2 (-1) dt = -2 \quad \tag{E2.18b.5}$$

$$\int_{C_2} \frac{-yu_x + xu_y}{x^2} \cdot d\mathbf{l} = \int_0^1 \frac{-u_x + (1+t)u_y}{(1+t)^2} \cdot \{(1+t)'u_x\}dt = \int_0^1 \frac{-1}{(1+t)^2} dt = \left[\frac{1}{1+t}\right]_0^1 = \frac{1}{2} - 1 = -\frac{1}{2} \quad \tag{E2.18b.6}$$


$$\int_{C_3} \frac{-yu_x + xu_y}{x^2} \cdot d\mathbf{l} = \int_0^2 \frac{-(1-t)u_x + 2u_y}{2^2} \cdot \{(1-t)'u_y\}dt = \int_0^2 \frac{1}{2} dt = -1 \quad \tag{E2.18b.7}$$


$$\int_{C_4} \frac{-yu_x + xu_y}{x^2} \cdot d\mathbf{l} = \int_0^1 \frac{-u_x + (2-t)u_y}{(2-t)^2} \cdot \{(2-t)'u_x\}dt = \int_0^1 \frac{-1}{(2-t)^2} dt = \left[\frac{1}{t-2}\right]_0^1 = -1 - \left(-\frac{1}{2}\right) = -\frac{1}{2} \quad \tag{E2.18b.8}$$

$$\int_{C_1+C_2+C_3+C_4} \frac{-yu_x + xu_y}{x^2} \cdot d\mathbf{l} = \text{Sum of Eqs. (E2.18b.6,7,8)} = -\frac{1}{2} - 1 - \frac{1}{2} = -2 \quad \tag{E2.18b.9}$$

Kesetaraan nilai integral garis pada kedua lintasan mendukung bahwa integral garis dari 
**G**(x, y) bersifat independen terhadap lintasan.
